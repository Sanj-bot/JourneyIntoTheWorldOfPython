```python
nums=[1,2,3,4,5]
for num in nums:
  print(num)
```

```python
#break statement to come out of a loop
for num in nums:
  if num==3:
    print('Found 3!')
    break
  print(num)
```

```python
#Continue loop
for num in nums:
  if num==4:
    print("Found 4!!")
    continue
  print(num)
```

```python
#Loop within a loop
for num in nums:
  for letter in 'abc':
    print(num,letter)
```
