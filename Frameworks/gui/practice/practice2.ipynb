```python
from tkinter import *

root=Tk()
root.title('My Box GUI')
root.iconbitmap('data/rabbit.ico')
root.geometry("500x500")
root.configure(bg='pink')

vertical_slider1=Scale(root,from_=0 ,to=300,bg='grey',orient=VERTICAL)
vertical_slider1.grid(row=0,column=0)
vertical_slider2=Scale(root,from_=0 ,to=300,bg='grey',orient=VERTICAL)
vertical_slider2.grid(row=0,column=1)
vertical_slider3=Scale(root,from_=0 ,to=300,bg='grey',orient=VERTICAL)
vertical_slider3.grid(row=0,column=2)
vertical_slider4=Scale(root,from_=0 ,to=300,bg='grey',orient=VERTICAL)
vertical_slider4.grid(row=0,column=3)
vertical_slider5=Scale(root,from_=0 ,to=300,bg='grey',orient=VERTICAL)
vertical_slider5.grid(row=0,column=4)
vertical_slider6=Scale(root,from_=0,to=1,bg='red',orient=VERTICAL)
vertical_slider6.grid(row=0,column=5)



def value():
    
    ver_label=Label(root,text=vertical_slider1.get())
    ver_label.grid(row=1,column=0)
    ver_label=Label(root,text=vertical_slider2.get())
    ver_label.grid(row=1,column=1)
    ver_label=Label(root,text=vertical_slider3.get())
    ver_label.grid(row=1,column=2)
    ver_label=Label(root,text=vertical_slider4.get())
    ver_label.grid(row=1,column=3)
    ver_label=Label(root,text=vertical_slider5.get())
    ver_label.grid(row=1,column=4)
    ver_label=Label(root,text=vertical_slider6.get())
    ver_label.grid(row=1,column=5)

ver_button=Button(root,text='Click for value',command=value)
ver_button.grid(row=2,column=10)








root.mainloop()
```
